```python
x = df0.drop(['SeriousDlqin2yrs'], axis=1)

x.head()
```

```python
y = df0['SeriousDlqin2yrs']
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y,  test_size = 0.30)
```

```python
xtest.info()
```

```python
from sklearn import tree

clf = tree.DecisionTreeClassifier()

clf.fit(xtrain, ytrain)
```

```python
preddt_train = clf.predict(xtrain)
preddt_test = clf.predict(xtest)

print('Training DT performace ', metrics.accuracy_score(ytrain, preddt_train))
print('Testing DT performace ', metrics.accuracy_score(ytest, preddt_test))
```

```python
labels = ['Non-deafult', 'default']

print('Training set Classification Report')
print('\n \n', classification_report(ytrain, preddt_train, target_names=labels))

print('Testing set Classification Report')
print('\n \n', classification_report(ytest, preddt_test, target_names=labels))
```

```python
gnb = GaussianNB()

gnb.fit(xtrain, ytrain)
```

```python
preddt_train = gnb.predict(xtrain)
preddt_test = gnb.predict(xtest)

print('Training DT performace ', metrics.accuracy_score(ytrain, preddt_train))
print('Testing DT performace ', metrics.accuracy_score(ytest, preddt_test))
```

```python
labels = ['Non-deafult', 'default']

print('Training set Classification Report')
print('\n \n', classification_report(ytrain, preddt_train, target_names=labels))

print('Testing set Classification Report')
print('\n \n', classification_report(ytest, preddt_test, target_names=labels))
```
